```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import urllib
from urllib.request import urlopen
```

```python
url = "https://www.myntra.com/shirts?rawQuery=shirts"
res = requests.get(url)
print(res.status_code)
```

```python
soup = BeautifulSoup(res.content,'html.parser')
soup
```

```python
#Product Name
prod_list = []
prod_name = soup.find_all('div', class_='_4rR01T')
for i in prod_name:
    prod_list.append(i.text.strip())
```

```python
#Features
feature_list = []
features = soup.find_all('ul', class_='_1xgFaf')
for i in features:
    feature_list.append(i.text.strip())

#feature_list
```

```python
# Prices
price_list = []
prices = soup.find_all('div','_30jeq3 _1_WHN1')
for i in prices:
    price_list.append(i.text.strip())
#price_list
```

```python

# Rating
rating_list = []
ratings = soup.find_all('div', class_ = '_3LWZlK')
for i in ratings:
    rating_list.append(i.text.strip())
#rating_list
```
